```python
df = pd.read_csv('classif_train.csv')
df['text'] = df.apply(lambda x: x['title'] + ' ' + x['text'], axis=1)
words = pd.read_csv('new_df_1.csv')
count_0 = set(words.sort_values(['count_0'], ascending=False)['Unnamed: 0'][:200])
count_1 = set(words.sort_values(['count_1'], ascending=False)['Unnamed: 0'][:200])
count_2 = set(words.sort_values(['count_2'], ascending=False)['Unnamed: 0'][:200])
count_3 = set(words.sort_values(['count_3'], ascending=False)['Unnamed: 0'][:100])
count_4 = set(words.sort_values(['count_4'], ascending=False)['Unnamed: 0'][:100])
count_5 = set(words.sort_values(['count_5'], ascending=False)['Unnamed: 0'][:100])
count_6 = set(words.sort_values(['count_6'], ascending=False)['Unnamed: 0'][:100])
count_7 = set(words.sort_values(['count_7'], ascending=False)['Unnamed: 0'][:50])
count_8 = set(words.sort_values(['count_8'], ascending=False)['Unnamed: 0'][:50])
words = set()
words = words.union(count_0, count_1, count_2, count_3, count_4, count_5, count_6, count_7, count_8)
len(words)
```

```python
for i in words:
    df[i] = df['text'].apply(lambda x: int(i in x))
df = df.drop('title', 1)
df = df.drop('text', 1)
df = df.drop('id', 1)
X = df.drop('type', axis=1)
y = df['type']
logistic = LogisticRegression()
np.mean(cross_val_score(logistic, X, y, cv=5))
```
